```python
results = []

totalLength = 0
resultlength = 0

for filename in filtered_results:
    df2 = pd.read_json(filename, lines=True)
    df2.pop('level')
    fn = os.path.basename(filename).split(".")[0]    
    df2["file"] = fn
    resultlength += len(df2)
    results.append(df2)


results_complete = pd.concat(results, ignore_index=True)

results_complete["StartDate_DT"] = pd.to_datetime(results_complete["StartDate"], format=timeFormat)
results_complete["CommitResultDate_DT"] = pd.to_datetime(results_complete["CommitResultDate"], format=timeFormat)
results_complete["EndInteractionDate_DT"] = pd.to_datetime(results_complete["EndInteractionDate"], format=timeFormat)

display (results_complete)
```

```python
study = []
totalLength = 0

for filename in filtered_data:
    df = pd.read_csv(filename, sep=";")
    fn = os.path.basename(filename).split(".")[0] 

    df["Block"] = pd.to_numeric(df["Block"])
    df["Trial"] = pd.to_numeric(df["Trial"])

    df["file"] = fn
    totalLength += len(df)
    study.append(df)
    
print("totallines: " + str(totalLength))

data_complete = pd.concat(study, ignore_index=True)

data_complete['Date'] = pd.to_datetime(data_complete["Date"], format=timeFormat)
# sort by date (necessary as some log messages are written in correct time order)
data_complete = data_complete.sort_values(by=['Date'], ignore_index=True)

data_complete['SubStateChange'] = data_complete['ProgramSubState'].diff()
data_complete['StateChange'] = data_complete['ProgramState'].diff()

data_complete['TrialChange'] = data_complete['Trial'].diff()

# compute significant state changes
num_trials = data_complete[data_complete['TrialChange'] != 0]

display(num_trials)

data_complete['StateChangeMarker']  = '-'
mask = (data_complete['StateChange'] == 1) & (data_complete['ProgramState'] == 4)

data_complete.loc[mask, 'StateChangeMarker'] = 'START_BLOCK | START_EXPERIMENT'

mask = (data_complete['StateChange'] == -1) & (data_complete['ProgramState'] == 3)
data_complete.loc[mask, 'StateChangeMarker'] = 'START_BLOCK | START_TRAINING'


data_complete.loc[0, 'StateChangeMarker'] = 'START_BLOCK | START_TRAINING'
# .loc[pd.isna(data_complete['StateChangeMarker'])]] = 

mask = (data_complete['SubStateChange'] == -2) & (data_complete['StateChange'] == 0)
data_complete.loc[mask, 'StateChangeMarker'] = 'START_BLOCK'

mask = (data_complete['SubStateChange'] != 0) & (data_complete['ProgramSubState'] == 1)
data_complete.loc[mask, 'StateChangeMarker'] = 'START_INTERACTION'

mask = (data_complete['SubStateChange'] == 1) & (data_complete['ProgramSubState'] == 2)
data_complete.loc[mask, 'StateChangeMarker'] = 'START_INPUT'

allProbands = data_complete['ProbandId'].drop_duplicates()

count_input = data_complete['StateChangeMarker'].eq('START_INPUT').sum()
count_interaction =  data_complete['StateChangeMarker'].eq('START_INTERACTION').sum()

count_block = data_complete['StateChangeMarker'].str.contains('START_BLOCK').sum()

count_training = data_complete['StateChangeMarker'].str.contains('START_TRAINING').sum()
count_experiment = data_complete['StateChangeMarker'].str.contains('START_EXPERIMENT').sum()

# assert correct numbers: 
display(f'Number of Participants: {len(allProbands)}')

# should be: 92 x numProbands
display(f'Number of Input-Markers: {count_input} [{len(allProbands)} x ((4 x 2) + (4 x 21))]')

# should be: 92 x numProbands
display(f'Number of Interaction-Markers: {count_interaction} [{len(allProbands)} x ((4 x 2) + (4 x 21))]')

# should be 8 x numProbands
display(f'Number of Blocks: {count_block} [{len(allProbands)} x (4 + 4)]')

# should be numProbands
display(f'Number of Training: {count_training}')

# should be numProbands
display(f'Number of Experiment: {count_experiment}')

data_complete.drop(df.columns[41], axis=1, inplace=True)

display (data_complete)
```

# remove leading and trailing idle times in interaction phase

```python
grouped = data_complete[data_complete['Block'] >= 0].groupby(['ProbandId', 'TrialNumber'])

results_complete['TrialNumber'] = 0
results_complete['CleanedUpInteractionStart_DT'] = results_complete['StartDate_DT']
results_complete['CleanedUpInteractionEnd_DT'] = results_complete['EndInteractionDate_DT']

for name, group in grouped:

   p = name[0]
   t = name[1]

   interactionStart = group[group['PosZ'] != '-'].head(1)
   interactionEnd = group[group['PosZ'] != '-'].tail(1)

   date_start = interactionStart['Date']
   date_stop = interactionEnd['Date']

   b = interactionStart['Block'].values[0]
   tId = interactionStart['Trial'].values[0]

   data_complete.loc[date_start.index[0], 'StateChangeMarker'] = 'INTERACTION_START_CLEANED'
   data_complete.loc[date_stop.index[0], 'StateChangeMarker'] = 'INTERACTION_END_CLEANED'

   associatedResult = results_complete[(results_complete['ProbandId'] == p) & (results_complete['BlockId'] == b) & (results_complete['TrialId'] == tId)]
   
   # if participant data is not existing in both data and results
   if len(associatedResult) <= 0:
      continue
   
   idx = associatedResult.index[0]
   
   results_complete.loc[idx, 'TrialNumber'] = t
   results_complete.loc[idx, 'CleanedUpInteractionStart_DT'] = date_start.values[0]
   results_complete.loc[idx, 'CleanedUpInteractionEnd_DT'] = date_stop.values[0]
   

results_complete['ResultNumberDifference'] = results_complete['ProvidedResultNumber'] - results_complete['ExpectedResultNumber']
results_complete['ResultLayerDifference'] = results_complete['ProvidedResultLayer'] - results_complete['ExpectedResultLayer']

results_complete['ResultLayerCorrect'] = results_complete['ResultLayerDifference'] == 0
results_complete['ResultNumberCorrect'] = results_complete['ResultNumberDifference'] == 0

results_complete['CommitResultDate_DT'] = results_complete['CommitResultDate_DT'].apply(pd.to_datetime)
results_complete['EndInteractionDate_DT'] = results_complete['EndInteractionDate_DT'].apply(pd.to_datetime)
results_complete['StartDate_DT'] = results_complete['StartDate_DT'].apply(pd.to_datetime)

# compute duration of interaction phase
results_complete['DurationMS'] = (results_complete['EndInteractionDate_DT'] - results_complete['StartDate_DT']).dt.total_seconds()

# compute duration of interaction phase
results_complete['DurationMS_Cleaned'] = (results_complete['CleanedUpInteractionEnd_DT'] - results_complete['CleanedUpInteractionStart_DT']).dt.total_seconds()

# need to compute a grouping index for conditions
results_complete['countCondition'] = results_complete.groupby(['Condition']).cumcount()

display(results_complete)
display(data_complete)

saveData(data_complete)
saveResults(results_complete)
```

```python
data_complete['diff_NumberDisplay'] = data_complete['DataNumberDisplayed'].diff()
data_complete.loc[0, 'diff_NumberDisplay'] = 0.0

data_complete['LayerChange'] = data_complete['Layer'].replace('-', -1).astype(float).diff()
data_complete.loc[0, 'LayerChange'] = -1

show = data_complete[data_complete['diff_NumberDisplay'] > 0.0]

subStateChange = data_complete[data_complete['SubStateChange'] != 0]
hide = data_complete[data_complete['diff_NumberDisplay'] < 0.0]

# hide = []

for idx, item in show.iterrows():
    nextHide = hide[hide['Date'] > item['Date']].head(1)
    nextStateChange = subStateChange[subStateChange['Date'] > item['Date']].head(1)

    show.at[idx, 'nextStateChangeIdx'] = int(nextStateChange.index[0])
    show.at[idx, 'nextStateChangeDate'] = nextStateChange['Date'].values[0]
    show.at[idx, 'nextHideIdx'] = int(nextHide.index[0])
    show.at[idx, 'nextHideDate'] = nextHide['Date'].values[0]

show = show.rename_axis('show_index').reset_index()
show['hide_index'] = np.minimum(show['nextStateChangeIdx'], show['nextHideIdx'])
show['hide_date'] = np.minimum(show['nextStateChangeDate'], show['nextHideDate'])

display(show)

df_NumberDisplayed = pd.DataFrame()
df_NumberDisplayed['ShowIndex'] = show['show_index']
df_NumberDisplayed['ShowDate'] = pd.to_datetime(show['Date'])
df_NumberDisplayed['HideIndex'] = show['hide_index']
df_NumberDisplayed['HideDate'] = show['hide_date']
df_NumberDisplayed['Condition'] = show['Condition']
df_NumberDisplayed['TrialNumber'] = show['TrialNumber']
df_NumberDisplayed['Layer'] = show['Layer']
df_NumberDisplayed['ResultLayer'] = show['ResultLayer']
df_NumberDisplayed['BlockId'] = show['Block']
df_NumberDisplayed['ProbandId'] = show['ProbandId']

# df_NumberDisplayed = df_NumberDisplayed[:10]

lastLayerChange = list(map(getLastLayerChange, df_NumberDisplayed['ShowIndex']))

df_layerChange = pd.DataFrame(lastLayerChange, columns=['LastLayerChangeDate','LastLayerChangeIndex'])
df_NumberDisplayed = pd.concat([df_NumberDisplayed, df_layerChange], axis = 1)

df_NumberDisplayed['LastLayerChangeDate'] = pd.to_datetime( df_NumberDisplayed['LastLayerChangeDate'])

df_NumberDisplayed['Duration_ShowNumber'] = (df_NumberDisplayed['HideDate'] - df_NumberDisplayed['ShowDate']).dt.total_seconds()
df_NumberDisplayed['Duration_FindCenter'] = (df_NumberDisplayed['ShowDate'] - df_NumberDisplayed['LastLayerChangeDate']).dt.total_seconds()

df_NumberDisplayed_experiment = df_NumberDisplayed[df_NumberDisplayed['BlockId'] >= 0]

df_NumberDisplayed.to_csv(rf'{export_data}numberDisplayed_all.csv', sep= ";")
df_NumberDisplayed_experiment.to_csv(rf'{export_data}numberDisplayed_experiment.csv', sep= ";")


display(df_NumberDisplayed.describe()['Duration_ShowNumber'])
```
